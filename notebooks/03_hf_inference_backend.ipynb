# 03 Hugging Face Inference
Uses **`HF_TOKEN`** for remote inference (cost / rate limits apply).

If `HF_TOKEN` is unset, this notebook uses `FakeLLMBackend` so cells still run.

```python
from pathlib import Path
import sys
for base in [Path.cwd().resolve(), *Path.cwd().resolve().parents]:
    src = base / "src"
    if (src / "agent_rpg").is_dir():
        if str(src) not in sys.path:
            sys.path.insert(0, str(src))
        break
from agent_rpg.repo_root import find_repo_root
ROOT = find_repo_root()
print("Repository root:", ROOT)
import os
```

```python
from agent_rpg import SimulationEngine, load_scenario
from agent_rpg.backends.fake import FakeLLMBackend
from agent_rpg.backends.hf_inference import HuggingFaceInferenceBackend

scenario = load_scenario(ROOT / "examples" / "scenarios" / "minimal.yaml")
scenario.orchestration.max_rounds = 1
scenario.world.max_rounds = 1
scenario.orchestration.enable_thought_phase = False

if os.environ.get("HF_TOKEN"):
    backend = HuggingFaceInferenceBackend()
    print("Using HuggingFaceInferenceBackend")
else:
    backend = FakeLLMBackend(
        factory=lambda i, m: '{\"thought\":\"\",\"say\":\"no token demo\",\"directed_at\":null}',
    )
    print("HF_TOKEN not set; using FakeLLMBackend")

out = SimulationEngine(scenario).run(backend, output_dir=ROOT / "runs", run_id="nb03_hf")
print(out)
```
