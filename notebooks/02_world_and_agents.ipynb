# 02 World and agents
Validate YAML scenarios with Pydantic.

```python
from pathlib import Path
import sys

for base in [Path.cwd().resolve(), *Path.cwd().resolve().parents]:
    src = base / "src"
    if (src / "agent_rpg").is_dir():
        if str(src) not in sys.path:
            sys.path.insert(0, str(src))
        break

from agent_rpg.repo_root import find_repo_root

ROOT = find_repo_root()
from pydantic import ValidationError
from agent_rpg.loader import load_scenario
from agent_rpg.schemas.scenario import ScenarioConfig

print("Repository root:", ROOT)
```

```python
s = load_scenario(ROOT / "examples" / "scenarios" / "tavern.yaml")
print(s.world.title, len(s.agents))
```

```python
bad = {
    "world": {"scenario_id": "x", "title": "t", "max_rounds": 1},
    "agents": [{"agent_id": "a", "display_name": "A", "relationships": {"nobody": "x"}}],
}
try:
    ScenarioConfig.model_validate(bad)
except ValidationError as e:
    print("Caught validation error as expected")
```
